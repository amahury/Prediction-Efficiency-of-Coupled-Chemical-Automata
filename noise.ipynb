## Different flavors of noise

```python
samples = 10**5
```

### White

```python
y = cn.powerlaw_psd_gaussian(0, samples)
a = abs(np.fft.rfft(y/max(y)))
plt.plot(a)
```

### Pink

```python
y = cn.powerlaw_psd_gaussian(1, samples)
a = abs(np.fft.rfft(y))
plt.plot(a)
```

### Brown

```python
y = cn.powerlaw_psd_gaussian(2, samples)
a = abs(np.fft.rfft(y))
plt.plot(a)
```

### Blue

```python
y = cn.powerlaw_psd_gaussian(-1, samples)
a = abs(np.fft.rfft(y))
plt.plot(a)
```

### Violet

```python
y = cn.powerlaw_psd_gaussian(-2, samples)
a = abs(np.fft.rfft(y))
plt.plot(a)
```
